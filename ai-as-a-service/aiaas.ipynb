## AI as a Service(AIaas): OpenAI Api

```python
from openai import OpenAI
```

```python
# Import your API secret
from google.colab import userdata
api_key = userdata.get('OPENAI_API_KEY_RAG')
```

```python
client = OpenAI(api_key=api_key)

def ask_ai(prompt, temperature=1):
  completion = client.chat.completions.create(
      model = 'gpt-4o',
      messages=[
          {'role': 'system', 'content': 'You are a Sci_fi writer and poet'},
          {'role': 'user', 'content': prompt}
      ],
      temperature=temperature
  )

  return completion.choices[0].message.content
```

```python
print(ask_ai('Write a poem about a lonely robot exploring a futuristic city.'))
```

```python
print(ask_ai('Can advanced AI replace human creativity, intuition, or emotions-what is often called "the human touch"- in the future'))
```
